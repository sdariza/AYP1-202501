Dado un n, escribe sus dígitos de izquierda a derecha

```python
n = -1; coc = 0; exp = 0; numInv = 0
while n < 0:
  n = int(input("Ingresa n: "))
coc = n
while coc != 0:
  exp = exp + 1
  coc = coc // 10
print(f"El número tiene {exp} digitos")
exp = exp - 1
coc = n
while coc != 0:
  numInv = numInv + coc % 10 * 10 ** exp
  exp = exp - 1
  coc = coc // 10
print(f"El número invertido es: {numInv}")
coc = numInv
print("El número leído de izquierda a derecha es:")
while coc != 0:
  print(coc%10, end="|")
  coc = coc // 10
```
