````python
def calcularSumDigitos(num):
  sum = 0
  while num != 0:
    dig = num % 10
    if dig % 2 == 0:
      sum = sum + dig
    num = num // 10
  return sum
````

````python
def contDigitos(num):
  cont = 0
  while num != 0:
    cont = cont + 1
    num = num // 10
  return cont
````

````python
def invertirNumero(num):
  numInv = 0
  exp = contDigitos(num) - 1
  while num != 0:
    dig = num % 10
    numInv = numInv  + dig * 10 ** exp
    exp = exp - 1
    num = num // 10
  return numInv
````

````python
def isPrime(num):
  cont = 0
  div = 1
  while div <= num and cont <= 2:
    if num % div == 0:
      cont = cont + 1
    div = div + 1
  if cont == 2:
    return True
  else:
    return False
````

````python
if __name__ == '__main__':
  n = 0
  while n <= 0:
    n = int(input("Ingrese n: "))
  suma_digitos = calcularSumDigitos(n)
  print(f"La suma de los digitos pares de {n} es {suma_digitos}")
  if suma_digitos > 20:
    print(f"La suma de los digitos pares de {n} es mayor a 20")
    numInv = invertirNumero(n)
    print(f"El numero {n} invertido es {numInv}")
  else:
    print(f"La suma de los digitos pares de {n} es menor a 20")
    if isPrime(n):
      print(f"{n} es un numero primo")
    else:
      print(f"{n} no es un numero primo")
````
